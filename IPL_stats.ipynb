# Q.5

```python
df10 = pd.read_csv('matches.csv')
df11 = pd.DataFrame(df10)
df11['counts'] = 1
df12 = df11[['toss_winner','winner','counts']]
df12
```

```python
# win = 0
# lose = 0
wn = df12[df12['toss_winner'] == df12['winner']]
ls = df12[df12['toss_winner'] != df12['winner']]
win = wn['counts'].count()
lose = ls['counts'].count()
   
print("Teams winning toss and winning the match : ",win)
print("Teams losing toss and winning the match : ",lose)
```

```python
sit = ["Winning toss and winning the match","Losing toss and winning the match"]
wins = [win,lose]

colors = ["orange","blue"]
plt.pie(wins, labels=sit,explode = None ,colors=colors,autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Teams winning a match")
plt.show()
```
